```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url_link):
  return url_link.split("watch?v=")[-1]
```

```python
video_id = get_video_id("https://www.youtube.com/watch?v=7owOKKgT0ak")
print(video_id)
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id)
print(transcript)
```

```python
transcript_joined = " ".join([line['text'] for line in transcript])
print(transcript_joined)
```

```python
from transformers import pipeline

generator = pipeline('text-generation', model='gpt2')

def chat(prompt):
    result = generator(prompt, max_length=150, temperature=0.7)
    return result[0]['generated_text']
```

```python
answer = chat(prompt)
print(answer)
```
